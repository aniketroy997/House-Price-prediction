```python
import random
data=[]
random.seed(502)
ran=["Super built-up","Built-up","Plot","Carpet"]
for i in range(2518):
    data1=random.choice(ran)
    data.append(data1)

print("Area_type Random",data)
```

```python
dataframe=pd.DataFrame()
dataframe["Area_type"]=data[:]
print(dataframe)
```

```python
dataframe.to_csv('result.csv', index = False)
```
